```python
import pandas as pd
import plotly.plotly as py
import plotly.graph_objs as go
import numpy as np 
import csv 
import os

"""Import train.csv"""
def import_csv_labels():
    data_header = []
    feed_rate = []
    clamp_pressure = []
    condition = []
    machine_finale = []
    visual_inspect = []
    with open('/Users/benjaminadame/desktop/train.csv', 'r') as csvfile:
        file = csv.reader(csvfile)
        file = list(file)
        data_header = file[0]
        #Delete the first two columns 
        for row in file:
            del row [0]
        for row in file:
            del row [0]
        #Establish the feed_rate and the clamp pressure and the rest of the labels
        for row in file:
            feed_rate.append(row[0])
            clamp_pressure.append(row[1])
            condition.append(row[2])
            machine_finale.append(row[3])
            visual_inspect.append(row[4])
            
        del feed_rate[0]
        del clamp_pressure[0]
        del condition[0]
        del machine_finale[0]
        del visual_inspect[0]
        
        #convert to int
        for i in range(len(feed_rate)):
            feed_rate[i] = float(feed_rate[i])
            clamp_pressure[i] = float(clamp_pressure[i])
            
        return data_header,feed_rate,clamp_pressure,condition,machine_finale,visual_inspect
    
"""Importing all the worn tools datasets"""
def import_worn_csv():
    
    """Fetching all the csv files that are all from the worn tools that finalized machining"""
    # Fetch data from directory "/Users/benjaminadame/desktop/cncproj/experiment_01...18
    worn_data = []
    files = []
    path = '/Users/benjaminadame/desktop/cncproj'
    for filename in os.listdir(path):
        files.append('/Users/benjaminadame/desktop/cncproj' + '/' + filename)
    files.remove(files[0])
    d_h,f_r,c_p,c,m_f,v_i = import_csv_labels()
    experiments = 18
    for i in range(0,experiments):
        if c[i] == 'worn'and m_f[i] == 'yes':
            # Fetch data from directory "/Users/benjaminadame/desktop/cncproj/experiment_01...18
            #######################################################################
            with open(files[i], 'r') as csvfile:
                wd = csv.reader(csvfile)
                wd = list(wd)
                worn_data.append(wd)
            ########################################################################

                
if __name__ =="__main__":
    import_worn_csv()
            
    
    
    
    
    
    




            
        
            

    
    
    
    
    
            


            
        
    
```
